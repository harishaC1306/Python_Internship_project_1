```python
import math

def mortgage_calculator():
    print("\n=== Mortgage Calculator ===")
    try:
        principal = float(input("Enter loan amount ($): "))
        annual_rate = float(input("Enter annual interest rate (%): "))
        years = int(input("Enter loan term (years): "))

        monthly_rate = annual_rate / 100 / 12
        total_payments = years * 12


        monthly_payment = principal * (monthly_rate * (1 + monthly_rate)**total_payments) / ((1 + monthly_rate)**total_payments - 1)

        print(f"\nMonthly Payment: ${monthly_payment:,.2f}")
        print(f"Total Payment: ${monthly_payment * total_payments:,.2f}")
    except ValueError:
        print("Please enter valid numerical values")

def investment_calculator():
    print("\n=== Investment Return Calculator ===")
    try:
        initial = float(input("Enter initial investment ($): "))
        annual_rate = float(input("Enter expected annual return (%): "))
        years = float(input("Enter investment time horizon (years): "))


        rate = annual_rate / 100
        future_value = initial * (1 + rate)**years

        print(f"\nFuture Value: ${future_value:,.2f}")
        print(f"Total Profit: ${future_value - initial:,.2f}")
    except ValueError:
        print("Please enter valid numerical values")

def savings_goal_calculator():
    print("\n=== Savings Goal Calculator ===")
    try:
        goal = float(input("Enter savings goal ($): "))
        years = float(input("Enter time frame (years): "))
        annual_rate = float(input("Enter expected annual return (%): "))

        months = years * 12
        monthly_rate = annual_rate / 100 / 12

        monthly_savings = goal * monthly_rate / ((1 + monthly_rate)**months - 1)

        print(f"\nRequired Monthly Savings: ${monthly_savings:,.2f}")
        print(f"Total Contributions: ${monthly_savings * months:,.2f}")
    except ValueError:
        print("Please enter valid numerical values")

def income_tax_calculator():
    print("\n=== Income Tax Calculator ===")

    tax_brackets = [
        (9950, 0.10),
        (40525, 0.12),
        (86375, 0.22),
        (164925, 0.24),
        (209425, 0.32),
        (523600, 0.35),
        (float('inf'), 0.37)
    ]

    try:
        income = float(input("Enter annual income ($): "))
        deductions = float(input("Enter total deductions ($): "))

        taxable_income = max(0, income - deductions)
        tax = 0
        previous_bracket = 0

        for bracket, rate in tax_brackets:
            if taxable_income > previous_bracket:
                taxable_in_bracket = min(taxable_income, bracket) - previous_bracket
                tax += taxable_in_bracket * rate
                previous_bracket = bracket
            else:
                break

        print(f"\nTaxable Income: ${taxable_income:,.2f}")
        print(f"Estimated Tax Liability: ${tax:,.2f}")
    except ValueError:
        print("Please enter valid numerical values")

def main():
    while True:
        print("\n=== Financial Planning Toolkit ===")
        print("1. Mortgage Calculator")
        print("2. Investment Return Calculator")
        print("3. Savings Goal Calculator")
        print("4. Income Tax Calculator")
        print("5. Exit")

        choice = input("\nSelect an option (1-5): ")

        if choice == '1':
            mortgage_calculator()
        elif choice == '2':
            investment_calculator()
        elif choice == '3':
            savings_goal_calculator()
        elif choice == '4':
            income_tax_calculator()
        elif choice == '5':
            print("Thank you for using the Financial Planning Toolkit!")
            break
        else:
            print("Invalid option. Please try again.")

        input("\nPress Enter to continue...")

if __name__ == "__main__":
    main()
```
